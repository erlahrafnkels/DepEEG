## Listing all the individual files
They don't all have the same naming format.

```python
data_path = 'G:/My Drive/Skoli/DTU/Thesis/DepEEG/Data/tDCS_EEG_data/'
subject_folders = os.listdir(data_path)
txt_file_paths = []

for subject in subject_folders:
  path = data_path + subject
  subfolders = os.listdir(path)
  for p_folders in subfolders:
    str = path + '/' + p_folders
    pre_post = os.listdir(str)
    for file in pre_post:
      file_path = str + '/' + file
      txt_file_paths.append(file_path)

print(txt_file_paths)
```
